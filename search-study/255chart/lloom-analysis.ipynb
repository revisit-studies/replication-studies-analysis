```python
import os
import pandas as pd

def transcripts_to_dataframe(folder_path):
    """
    Reads transcripts from files in a folder and creates a pandas DataFrame.

    Args:
        folder_path (str): The path to the folder containing the transcript files.

    Returns:
        pandas.DataFrame: A DataFrame with 'name' (filename) and 'content' (transcript text) columns.
        Returns None if an error occurs or the folder is empty.
    """
    try:
        data = []
        for filename in os.listdir(folder_path):
            filepath = os.path.join(folder_path, filename)
            if os.path.isfile(filepath):  # Ensure it's a file, not a directory
                try:
                    with open(filepath, 'r', encoding='utf-8') as file:
                        content = file.read()
                        data.append({'name': filename, 'content': content})
                except Exception as e:
                    print(f"Error reading file {filename}: {e}")

        if not data:
            print("No transcripts found in the specified folder.")
            return None

        df = pd.DataFrame(data)
        return df

    except FileNotFoundError:
        print(f"Error: Folder not found at {folder_path}")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None

# Example usage:
folder_path_search = "transcripts-search"  
folder_path_control = "transcripts-control"
df_transcripts_search = transcripts_to_dataframe(folder_path_search)
df_transcripts_control = transcripts_to_dataframe(folder_path_control)
df_transcripts_search['isSearch'] = True
df_transcripts_control['isSearch'] = False
df_transcripts = pd.concat([df_transcripts_search, df_transcripts_control], ignore_index=True)
if df_transcripts is not None:
    display(df_transcripts)
```

```python
l = wb.lloom(
    df=df_transcripts,
    text_col="content",
    id_col="name"
)
```

```python
await l.gen(
    seed="search",  # Optional
)
```
